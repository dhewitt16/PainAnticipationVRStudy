```python
##Figures for pupil diameter

import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt

def export_fig(axis,text, fname):
    if savefigs:
        axis.text()
        axis.savefig(fname, bbox_inches='tight')

measureType = "GazeDirection" # GazeDirection or PupilDiameter
measureTypeLabel = "Gaze Direction"
directional = True

path = "/Users/dhewitt/Data/pps/Exports/"
fileType = ".csv"
fileName = ("Combined_" + measureType + "_noav_1904")
fileNameExt = ("Combined_" + measureType + "Ext_1904")
fullName = (path + measureType + "/" + fileName + fileType)
fullExtName = (path + measureType + "/" + fileNameExt + fileType)
figs_dir = (path + 'figs/')
print(fullName)
        
savefigs = True
if savefigs:
    # Make the figures folder if it doesn't yet exist
    if not os.path.isdir(figs_dir):
        os.makedirs(figs_dir)
```

```python
figSize = 3
timesSmaller = 8/figSize
#lW = 2.66/timesSmaller #indiivdual datapoint line width
lW = 2.4/timesSmaller #indiivdual datapoint line width
#eW = 3.57/timesSmaller #bar chart line width
eW = 1.6/timesSmaller #bar chart line width
spineW = 1/timesSmaller
sns.set(style="white",font_scale=2.8/timesSmaller)
palette = sns.color_palette('Set2')
```

```python

## Pupil Diameter
data2load = pd.read_csv(fullName)

df = pd.DataFrame(data2load)
#df = pd.DataFrame(data2load).drop(['Timepoint','ID'], axis=1)
df=df.rename(columns={'Unnamed: 0':'Timepoint', 'Participant':'ID'})
df.head()
```

```python
conditions = [
    (df['Timepoint'] >= 0) & (df['Timepoint'] < 800),
    (df['Timepoint'] >= 800) & (df['Timepoint'] < 1800),
    (df['Timepoint'] >= 1800) & (df['Timepoint'] < 2800),
    (df['Timepoint'] >= 2800) & (df['Timepoint'] < 3300)
]
values = [1, 2, 3, 4]  # Corresponding values for each condition

# Create the 'phase' column based on conditions and values
df['Phase'] = np.select(conditions, values, default=9)  # Use default value for unmatched conditions

df=df.rename(columns={'Unnamed: 0':'Timepoint', 'Participant':'ID'})
df
```

```python
avdf = df.groupby(['Timepoint']).mean().drop('ID',axis=1)

avdf['CM'] = avdf[['CM1', 'CM2']].mean(axis=1)
avdf['CR'] = avdf[['CR1', 'CR2']].mean(axis=1)
avdf['CL'] = avdf[['CL1', 'CL2']].mean(axis=1)
avdf['EM'] = avdf[['EM1', 'EM2']].mean(axis=1)
avdf['ER'] = avdf[['ER1', 'ER2']].mean(axis=1)
avdf['EL'] = avdf[['EL1', 'EL2']].mean(axis=1)

newdf = avdf.copy()
newdf['CP'] = newdf[['CL', 'CR']].mean(axis=1)
newdf['EP'] = newdf[['EL', 'ER']].mean(axis=1)
```

```python

f, ax = plt.subplots(figsize=(figSize*1.6, figSize)) # for a flatter, wider chart

plt.axvline(x=800, color='Black', lw=eW)
plt.axvline(x=1800, color='Black', linestyle='--', lw=eW)
plt.axvline(x=2800, color='Black', linestyle='--', lw=eW)

# Using shaded areas
#plt.axvspan(0, 800, color='orange', alpha=0.10, lw=0)
#plt.axvspan(800, 1800, color='green', alpha=0.10, lw=0)
#plt.axvspan(1800, 2800, color='blue', alpha=0.10, lw=0)
#plt.axvspan(2800, 3300, color='gray', alpha=0.10, lw=0)

plt.plot(newdf['CM'],label="Neutral Conditioning", color=palette[0], lw = lW)
plt.plot(newdf['CP'],label="Pain Conditioning", color=palette[1], lw = lW)
plt.plot(newdf['EM'],label="Neutral Extinction", color=palette[0], linestyle='--', lw = lW)
plt.plot(newdf['EP'],label="Pain Extinction", color=palette[1], linestyle='--', lw = lW)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines['bottom'].set_color('Black')
ax.spines['bottom'].set_linewidth(eW)
ax.spines['left'].set_color('Black')
ax.spines['left'].set_linewidth(eW)
ax.tick_params(axis='both', which='both', pad=5) #change pad to make the axes labels further/closer to axis
ax.xaxis.labelpad = 3 # space between axis label and data name
ax.yaxis.labelpad = 1
ax.set_xlabel('Time (s)')
ax.set_ylabel(measureTypeLabel + ' (mms)') 

num_ticks = 6  # Adjust as needed
tick_locations = np.linspace(0, 3000, num_ticks)  # Adjust the range as needed
plt.xticks(tick_locations)

ax = plt.gca()
ax.set_ylim([-3, 3]) #for gazedir only

#plt.legend(loc='center left', bbox_to_anchor=(1, 0.5)) #uncomment to show legend

if savefigs:
    plt.savefig(figs_dir + fileName +"_fig3sized.svg", format='svg', dpi=1200, bbox_inches='tight')

plt.show()
```
