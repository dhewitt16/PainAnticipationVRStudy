```python
data2load = pd.read_csv('/Users/dhewitt/Data/pps/tonicpainsides.csv')
sidesdf = pd.DataFrame(data2load).astype('int32')
```

```python
melted_df = pd.merge(df,sidesdf, on='ID')
melted_df.drop(columns=['B2'],inplace = True)
melted_df
```

```python
##new version

# For cong DataFrame
cong_index = ((melted_df['Rep'].astype(int)==1) & (melted_df['B1'].astype(int)==1) & 
              (melted_df['Side']=='L')) | \
             ((melted_df['Rep'].astype(int)==1) & (melted_df['B1'].astype(int)==2) & 
              (melted_df['Side']=='R')) | \
             ((melted_df['Rep'].astype(int)==2) & (melted_df['B1'].astype(int)==1) & 
              (melted_df['Side']=='R')) | \
             ((melted_df['Rep'].astype(int)==2) & (melted_df['B1'].astype(int)==2) & 
              (melted_df['Side']=='L'))
cong = melted_df.loc[cong_index].copy()  # Ensure a copy is made to avoid modifying the original DataFrame
cong['congruency'] = 2

incong_index = ((melted_df['Rep'].astype(int) == 1) & (melted_df['B1'].astype(int) == 1) & 
                (melted_df['Side'] == 'R')) | \
               ((melted_df['Rep'].astype(int) == 1) & (melted_df['B1'].astype(int) == 2) & 
                (melted_df['Side'] == 'L')) | \
               ((melted_df['Rep'].astype(int) == 2) & (melted_df['B1'].astype(int) == 1) & 
                (melted_df['Side'] == 'L')) | \
               ((melted_df['Rep'].astype(int) == 2) & (melted_df['B1'].astype(int) == 2) & 
                (melted_df['Side'] == 'R'))
incong = melted_df.loc[incong_index].copy()  # Ensure a copy is made to avoid modifying the original DataFrame
incong['congruency'] = 1

neutral_index = melted_df['Side'] == 'M'
neutral = melted_df.loc[neutral_index].copy()  # Ensure a copy is made to avoid modifying the original DataFrame
neutral['congruency'] = 0

joineddf=pd.concat([neutral,cong,incong])
joineddf
joineddf_sorted = joineddf.sort_values(by='EEGPowerChange')
joineddf_sorted
```

```python
joineddf=joineddf.replace({'L':'1','M':'0','R':'2'})
joineddf.to_csv('/Users/dhewitt/Data/pps/Exports/ERD/PPSERDDataLong_Grouped_noav_180324.csv')
joineddf
```

# Adding VR data

```python
## Loading VR Data now - keeping phase column and renaming to timebin to match EEG data

data2load = pd.read_csv('/Users/dhewitt/Data/pps/Exports/PupilDiameter/GrandAvExport_PupilDiameterExtLong_Recoded_phaseav.csv')
pupilDiameter = pd.DataFrame(data2load)

data2load = pd.read_csv('/Users/dhewitt/Data/pps/Exports/GazeDirection/GrandAvExport_GazeDirExtLong_Recoded_phaseav.csv')
gazeDirection = pd.DataFrame(data2load)

VRdf = pd.merge(pupilDiameter, gazeDirection, on=['ID','Phase','Side','Rep','Condition'])
VRdf = VRdf.rename({'Measure_x' : 'PupilDiameter',
                    'Measure_y' : 'GazeDirection',
                   'Phase' : 'Timebin', 'Condition' : 'Block'}, axis=1)
#VRdf['Timebin'].replace({2:1,3:2}, inplace=True)
VRdf= VRdf.replace({'L':'1','M':'0','R':'2','C':'1','E':'2'})
VRdf = VRdf.astype('float64') # convert to numeric data types
```

```python
new_result_df = pd.merge(result_df, VRdf, on=['ID', 'Timebin', 'Side','Rep','Block'])
new_result_df
```

# Now sorting the data by contra - ipsi

```python
cue_left = new_result_df[new_result_df['Side'] == 1]
cue_right = new_result_df[new_result_df['Side'] == 2]
```

```python
cue_left_contra = cue_left[cue_left['ElSide'] == 1].copy()
cue_left_ipsi = cue_left[cue_left['ElSide'] == -1].copy()
cue_left_contra = cue_left_contra.reset_index(drop=True)
cue_left_ipsi = cue_left_ipsi.reset_index(drop=True)
cue_left_contra_ratio = cue_left_contra.copy()
cue_left_contra_ratio['CIRatio'] = (cue_left_contra['EEGPowerChange'] - cue_left_ipsi['EEGPowerChange'])/(cue_left_contra['EEGPowerChange'] + cue_left_ipsi['EEGPowerChange'])
cue_left_contra_ratio
```

```python
cue_right_contra = cue_right[cue_right['ElSide'] == -1].copy()
cue_right_ipsi = cue_right[cue_right['ElSide'] == 1].copy()
cue_right_contra = cue_right_contra.reset_index(drop=True)
cue_right_ipsi = cue_right_ipsi.reset_index(drop=True)
cue_right_contra_ratio = cue_right_contra.copy()
cue_right_contra_ratio['CIRatio'] = (cue_right_contra['EEGPowerChange'] - cue_right_ipsi['EEGPowerChange'])/(cue_right_contra['EEGPowerChange'] + cue_right_ipsi['EEGPowerChange'])
```

```python
new_result_df
```

```python
contra_ipsi_df = pd.concat([cue_left_contra_ratio,cue_right_contra_ratio])
```
