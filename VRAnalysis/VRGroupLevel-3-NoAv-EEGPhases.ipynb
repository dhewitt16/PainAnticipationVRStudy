```python
### this code takes data which was not averaged 
## over each run/repetition of block,
## so rep can be added into the LME

## updated on 4th march to add the covariates of interest

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

## Pupil Diameter
#data2load = pd.read_csv('/Users/dhewitt/Data/pps/Exports/PupilDiameter/Combined_PupilDiameter_noav.csv')

## Gaze Direction
data2load = pd.read_csv('/Users/dhewitt/Data/pps/Exports/GazeDirection/Combined_GazeDirection_noav.csv')

## Gaze Object / Fixation
#data2load = pd.read_csv('/Users/dhewitt/Data/pps/Exports/Combined_GazeObjectData_noav.csv')

## Hand position
#data2load = pd.read_csv('/Users/dhewitt/Data/pps/Exports/allHandData_2911.csv')
```

```python
df = pd.DataFrame(data2load)
#df = pd.DataFrame(data2load).drop(['Timepoint','ID'], axis=1)
df=df.rename(columns={'Unnamed: 0':'Timepoint', 'Participant':'ID'})
df.head()
```

```python
conditions = [
    (df['Timepoint'] >= 0) & (df['Timepoint'] < 800),
    (df['Timepoint'] >= 800) & (df['Timepoint'] < 1800),
    (df['Timepoint'] >= 1800) & (df['Timepoint'] < 2800),
    (df['Timepoint'] >= 2800) & (df['Timepoint'] < 3300)
]
values = [1, 2, 3, 4]  # Corresponding values for each condition

# Create the 'phase' column based on conditions and values
df['Phase'] = np.select(conditions, values, default=9)  # Use default value for unmatched conditions

# Display the updated DataFrame
df
```

```python
list(df.columns)
df
```

```python
melted_df = pd.melt(df, id_vars = ['Timepoint','Phase','ID'], value_vars=['CM1', 'CL1', 'CR1', 'EM1', 'EL1', 'ER1', 'CM2', 'CL2', 'CR2', 'EM2', 'EL2', 'ER2'])
melted_df[['Condition', 'Side', 'Rep']] = melted_df['variable'].str.extract(r'([CEMLR])([CEMLR])([12])')
melted_df = melted_df.drop(columns=['variable'])
melted_df.rename({'value':'Measure'}, axis=1, inplace=True)
melted_df
```

```python
melted_df = melted_df.reset_index()
melted_df
```

```python
# Averaging for each phase
average_df = melted_df.groupby(['ID', 'Phase', 'Side', 'Rep', 'Condition']).mean().reset_index()
average_df.drop(['Timepoint','index'],axis=1, inplace=True)
average_df
```

```python
melted_df = average_df

data2load = pd.read_csv('/Users/dhewitt/Data/pps/tonicpainsides.csv')
sidesdf = pd.DataFrame(data2load).astype('int32')
melted_df=melted_df.join(sidesdf.set_index('ID'), on='ID')
melted_df.drop(columns=['B2'],inplace = True)
melted_df
```

```python
# For cong DataFrame
cong_index = ((melted_df['Rep'].astype(int)==1) & (melted_df['B1'].astype(int)==1) & 
              (melted_df['Side']=='L')) | \
             ((melted_df['Rep'].astype(int)==1) & (melted_df['B1'].astype(int)==2) & 
              (melted_df['Side']=='R')) | \
             ((melted_df['Rep'].astype(int)==2) & (melted_df['B1'].astype(int)==1) & 
              (melted_df['Side']=='R')) | \
             ((melted_df['Rep'].astype(int)==2) & (melted_df['B1'].astype(int)==2) & 
              (melted_df['Side']=='L'))
cong = melted_df.loc[cong_index].copy()  # Ensure a copy is made to avoid modifying the original DataFrame
cong['congruency'] = 2

incong_index = ((melted_df['Rep'].astype(int) == 1) & (melted_df['B1'].astype(int) == 1) & 
                (melted_df['Side'] == 'R')) | \
               ((melted_df['Rep'].astype(int) == 1) & (melted_df['B1'].astype(int) == 2) & 
                (melted_df['Side'] == 'L')) | \
               ((melted_df['Rep'].astype(int) == 2) & (melted_df['B1'].astype(int) == 1) & 
                (melted_df['Side'] == 'L')) | \
               ((melted_df['Rep'].astype(int) == 2) & (melted_df['B1'].astype(int) == 2) & 
                (melted_df['Side'] == 'R'))
incong = melted_df.loc[incong_index].copy()  # Ensure a copy is made to avoid modifying the original DataFrame
incong['congruency'] = 1

neutral_index = melted_df['Side'] == 'M'
neutral = melted_df.loc[neutral_index].copy()  # Ensure a copy is made to avoid modifying the original DataFrame
neutral['congruency'] = 0

joineddf=pd.concat([neutral,cong,incong])
joineddf
```

```python
joineddf=joineddf.replace({'C':'1', 'E':'2','M':'0','L':'1','R':'2'})
joineddf['Side'] = joineddf['Side'].astype('int64')
joineddf['Rep'] = joineddf['Rep'].astype('int64')
joineddf
```

```python
result_df = pd.merge(joineddf, ratingsdf, on=['ID', 'B1', 'Rep'])
result_df
```
